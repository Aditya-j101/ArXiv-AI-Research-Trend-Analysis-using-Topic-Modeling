```python
import re
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

##Text PreProcessing

```python
import nltk
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('punkt')
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))
def preprocess_text(text):
  text = re.sub(r'[^a-zA-Z\s]', ' ', text.lower())
  tokens = word_tokenize(text)
  processed_tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words and len(word) > 3]
  return " ".join(processed_tokens)

```

```python
print("Starting Step 1: Text Preprocessing...")
# Apply the cleaning function to the 'text' column of your recent papers
df_recent['processed_text'] = df_recent['text'].apply(preprocess_text)
print("✅ Text Preprocessing complete.")
print(df_recent[['text', 'processed_text']].head())
```

## Feature Engineering
max_df ignore the words that are too common.

min_df ignore the words that are too rare.

```python
print("Step 2: Starting Feature Engineering")
vectorizer = CountVectorizer(max_df=0.95, min_df=5, stop_words='english')
doc_term_matrix = vectorizer.fit_transform(df_recent['processed_text'])
print("✅ Document-Term Matrix created successfully.")
print(f"Matrix Shape: {doc_term_matrix.shape} (Papers, Unique Words)")
```

## LDA Modelling

```python
import pyLDAvis
num_topics = 10
lda_model = LatentDirichletAllocation(n_components = num_topics,
                                      random_state = 42)
print(f"Training LDA model with {num_topics} topics...")
lda_model.fit(doc_term_matrix)
print("LDA Model training Complete")
```

```python
topic_term_dists = lda_model.components_
doc_topic_dists = lda_model.transform(doc_term_matrix)
doc_lengths = doc_term_matrix.sum(axis=1).A1
vocab = vectorizer.get_feature_names_out()
term_frequency = doc_term_matrix.sum(axis=0).A1
vis_data = pyLDAvis.prepare(
    topic_term_dists,
    doc_topic_dists,
    doc_lengths,
    vocab,
    term_frequency
)

pyLDAvis.display(vis_data)
```

```python
topic_labels = {0:"AI Reasoning and Search",
                1:"Graph Neural Networks and Embeddings",
                2:"AI Ethics and HealthCare",
                3:"Explainable AI for Large Language Models",
                4:"Deep Reinforcement Learning for Games",
                5:"Knowledge Representation and Reasoning",
                6:"Decision Making under Uncertatinity",
                7:"Time Series and Analysis and Data Mining",
                8:"Automated Planning and Robotics",
                9:"Autonomous Driving and Traffic Optimization"}
```

```python
doc_topic_dists = lda_model.transform(doc_term_matrix)
df_recent['dominant_topic'] = np.argmax(doc_topic_dists, axis=1)
df_recent['topic_label'] = df_recent['dominant_topic'].map(topic_labels)

print("Assigned dominant topic to each paper:")
print(df_recent[['year', 'dominant_topic', 'topic_label']].tail())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Group by year and topic label, then get the count and normalize
trends = df_recent.groupby('year')['topic_label'].value_counts(normalize=True).unstack()

# Plotting the trends
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(15, 8))

# Plot a line for each topic
trends.plot(ax=ax, linewidth=2.5)

# Formatting the plot
ax.set_title('Trends of AI Research Topics on ArXiv (2010-2025)', fontsize=16)
ax.set_ylabel('Proportion of Papers', fontsize=12)
ax.set_xlabel('Year', fontsize=12)
ax.legend(title='Topic', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
